## Problem Statement:
### Consider any image with size 1024x1024. Modify the image to the sizes 512x512, 256x256, 128x128, 64x64 and 32x32 using subsampling technique. Create the original image from all the above subsampled images using resampling technique.

```python
%matplotlib inline

from matplotlib import pyplot as plt
import cv2
```

```python
window_name = 'image'

# Load the image
image = cv2.imread("1138973.webp")
plt.imshow(image)
plt.show()
```

```python
height, width = image.shape[:2]

print(width)
print(height)
```

### Sub-sampling the original image (1024x1024) to size: 512x512.

```python
# Calculate the new dimensions of the image
new_height = 512
new_width = 512

# Resize the image using subsampling
resized_image1 = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
plt.imshow(resized_image1)
plt.show()
```

```python
# Calculate the new dimensions of the image
new_height = 256
new_width = 256

# Resize the image using subsampling
resized_image2 = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
plt.imshow(resized_image2)
plt.show()
```
